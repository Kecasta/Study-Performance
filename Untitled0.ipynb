**RENDIMIENTO DE ESTUDIANTES**

**1. Importamos la base de datos**

```python
import pandas as pd # importamos panda para la lectura de la base de datos
```

```python
students_df = pd.read_csv('/content/study_performance.csv') # leemos la base de datos
students_df # visualizamos la base de datos
```

**2. Pre-procesamiento**

2.1 Vamos a calcular una variable que nos sirva como una media a la cual vamos a llamar Score


```python
score_columns = [columna for columna in students_df.columns if columna.endswith('_score')] # escogemos las columnas a usar
score_columns
```

```python
students_df['Score'] = round(students_df[score_columns].sum(axis=1)/30) # vamos a calcular una columna nueva llamada score, redondeamos las columnas sumandolas y dividiendo por 30
students_df
```

```python
students_df.drop(columns = score_columns, inplace = True) # eliminamos las columnas que no nos sirven
students_df
```

2.2 Variables categoricas

  vamos a transformar las columnas de texto en columnas numericas usando un One hot encoding por ejemplo en las columna de genero vamos a hacer dos columnas female y male una con valor 0 y la otra con valor 1

```python
from sklearn.preprocessing import OneHotEncoder # importamos OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output=False) # preparamos el encoder
```

```python
encoded_data = encoder.fit_transform(students_df[['gender']]) # transformamos genero a numerico
```

```python
encode_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['gender'])) # creamos un dataframe con los datos transformados
encode_df
```

```python
students_df = pd.concat([students_df, encode_df], axis=1) # concatenamos los datos transformados con los datos originales
students_df

students_df.drop(columns = ['gender'], inplace = True) # eliminamos la columna original
students_df
```

```python
encoder = OneHotEncoder(sparse_output=False) # preparamos el encoder
encoded_data = encoder.fit_transform(students_df[['test_preparation_course']]) # transformamos a numerico
encode_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['test_preparation_course'])) # creamos un dataframe con los datos transformados
students_df = pd.concat([students_df, encode_df], axis=1) # concatenamos los datos transformados con los datos originales
students_df.drop(columns = ['test_preparation_course'], inplace = True) # eliminamos la columna original
students_df
```

```python
encoder = OneHotEncoder(sparse_output=False) # preparamos el encoder
encoded_data = encoder.fit_transform(students_df[['lunch']]) # transformamos a numerico
encode_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['lunch'])) # creamos un dataframe con los datos transformados
students_df = pd.concat([students_df, encode_df], axis=1) # concatenamos los datos transformados con los datos originales
students_df.drop(columns = ['lunch'], inplace = True) # eliminamos la columna original
students_df
```

```python
encoder = OneHotEncoder(sparse_output=False) # preparamos el encoder
encoded_data = encoder.fit_transform(students_df[['race_ethnicity']]) # transformamos a numerico
encode_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['race_ethnicity'])) # creamos un dataframe con los datos transformados
students_df = pd.concat([students_df, encode_df], axis=1) # concatenamos los datos transformados con los datos originales
students_df.drop(columns = ['race_ethnicity'], inplace = True) # eliminamos la columna original
students_df
```

2.3 Ordinal encoding

```python
students_df['parental_level_of_education'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder # importamos OrdinalEncoder
ordinal_categories = ['some high school','high school','some college','associate\'s degree','bachelor\'s degree','master\'s degree'] # creamos las categorias
encoder = OrdinalEncoder(categories=[ordinal_categories]) # creamos el encoder
encoded_data = encoder.fit_transform(students_df[['parental_level_of_education']]) # transformamos a numerico
df_encoded = pd.DataFrame(encoded_data, columns=['parental_level_of_education_encoded']) # creamos un dataframe con los datos transformados
students_df = pd.concat([students_df, df_encoded], axis=1) # concatenamos los datos transformados con los datos originales
students_df.drop(columns = ['parental_level_of_education'], inplace = True) # eliminamos la columna original
students_df
```

**3. Hacemos nuestro modelo de Machine Learning**

```python
from sklearn.model_selection import train_test_split # importamos train_test_split
from sklearn.ensemble import RandomForestRegressor # importamos RandomForestRegressor
from sklearn.linear_model import LinearRegression # importamos LinearRegression
from sklearn.metrics import mean_squared_error # importamos mean_squared_error
```

```python
X = students_df.drop(columns = ['Score']) # definimos X
y = students_df['Score'] # definimos y

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21) # dividimos los datos en entrenamiento y prueba
X_train
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42) # creamos el modelo
rf_model.fit(X_train, y_train) # lo entrenamos

lr_model = LinearRegression() # creamos el modelo
lr_model.fit(X_train, y_train) # lo entrenamos
```

```python
rf_predictions = rf_model.predict(X_test) # hacemos las predicciones
lr_predictions = lr_model.predict(X_test) # hacemos las predicciones
```

```python
rf_mse = mean_squared_error(y_test, rf_predictions) # calculamos el error
lr_mse = mean_squared_error(y_test, lr_predictions) # calculamos el error

print("Random Forest MSE:", rf_mse) # lo imprimimos
print("Linear Regression MSE:", lr_mse) # lo imprimimos
```
